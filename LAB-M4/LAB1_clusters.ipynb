# ESCALADO DE DATOS

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(data)
```

# METODO DEL CODO

```python
from sklearn.cluster import KMeans
```

```python
inertia = []
K = range(1, 10)

for k in K:
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(X_scaled)
    inertia.append(model.inertia_)

plt.plot(K, inertia, 'o-')
```

# K-MEANS

```python
kmeans = KMeans(n_clusters=3, random_state=42)
data['cluster'] = kmeans.fit_predict(X_scaled)

data.head()
```

```python
data['cluster'].value_counts()
```

# ANALISIS DE COMPONENTES PRINCIPALES (PCA)

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# TRANSFORMAR A DATAFRAME DE PANDAS

pca_df = pd.DataFrame(X_pca, columns=['PCA1', 'PCA2'])
pca_df['cluster'] = data['cluster']
```

```python
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c=pca_df['cluster'], cmap='viridis')
```

```python
from sklearn.metrics import silhouette_score
silueta_kmeans = silhouette_score(X_scaled, data['cluster'])
silueta_kmeans
```

# CLUSTER JERARQUICO

```python
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
```

```python
linked = linkage(X_scaled, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked, truncate_mode='lastp', p=30, leaf_rotation=90, leaf_font_size=10)
plt.title('Dendrograma de Clustering Jerrquico')
plt.xlabel('Clientes agrupados')
plt.ylabel('Distancia (Ward)')
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster

clusters_h = fcluster(linked, t=2, criterion='maxclust')
```

```python
from scipy.cluster.hierarchy import fcluster

clusters_h = fcluster(linked, t=30, criterion='distance')
```

```python
data['cluster_h'] = clusters_h
data.head()
```

```python
data['cluster_h'].value_counts()
```

```python
pca_df['cluster_h'] = data['cluster_h'].values

plt.figure(figsize=(8,6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], 
            c=pca_df['cluster_h'], 
            cmap='viridis', 
            s=40)
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Clusters Jerrquicos')
plt.show()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
```

```python
for e in [0.8, 1.0, 1.2, 1.5, 2.0]:
    cluster_db = DBSCAN(eps=e, min_samples=5).fit_predict(X_scaled)
    print(e, len(set(cluster_db)) - (1 if -1 in cluster_db else 0))
```

```python
data['cluster_db'] = cluster_db
```

```python
data['cluster_db'].value_counts()
```

```python
pca_df['cluster_db'] = data['cluster_db'].values

plt.figure(figsize=(8,6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'],
            c=pca_df['cluster_db'],
            cmap='viridis',
            s=40)

plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Clusters DBSCAN)')
plt.show()
```

```python
print("K-Means:", data['cluster'].nunique(), "clusters")
print("Jerrquico:", data['cluster_h'].nunique(), "clusters")
print("DBSCAN:", data['cluster_db'].nunique(), "clusters")
```

# SILHOUETTE SCORE

```python
from sklearn.metrics import silhouette_score

print("Silhouette K-Means:", silhouette_score(X_scaled, data['cluster']))
print("Silhouette Jerrquico:", silhouette_score(X_scaled, data['cluster_h']))
```

```python
mask = data['cluster_db'] != -1
silhouette_score(X_scaled[mask], data.loc[mask, 'cluster_db'])
```
